Matplotlib

matplotlib.pyplot is a collection of command style functions that make matplotlib work like MATLAB and used for ploting graphs and displaying images.

Matplotlib is a multi-platform data visualization library built on NumPy arrays

Matplotlib consists of several plots like line, bar, scatter, histogram etc.

Whenever we plot a graph, the axes adjust and take the default ticks and can be customized.

```python
import matplotlib.pyplot as plt
import numpy as np
x=np.array([1,2,3,4,5,6,7])
y=np.exp(x)
plt.plot(x, y,label='Exponential',color='r',linewidth=1) # you can change the colors and other parameters
plt.xlabel('x')
plt.ylabel('y=exp(x)')
plt.legend()
plt.show()
```

```python
x=np.array([1,2,3,4,5,6,7])
y=np.exp(x)
plt.plot(x, y,'go',label='Square') # you can change the colors and other parameters
plt.xlabel('x')
plt.ylabel('y=x**2')
plt.legend()
plt.plot(x, y,'r--')
plt.show()
```

```python
x=np.arange(0,5,1)
print(x)
#plt.plot(x,np.exp(x),'r--',x,x**2,'bs',label='Exponential')
plt.plot(x,np.exp(x),'r--',label='Exponential')
plt.plot(x,x**2,'b--',label='Square')
plt.legend()
plt.xlabel('Numbers')
plt.ylabel('Output')
plt.title('Exp and Sq')
plt.show()
```

```python
import matplotlib.pyplot as plt
days = [1,2,3,4,5]
sleeping =[7,8,6,11,7]
eating = [2,3,4,3,2]
working =[7,8,7,2,2]
playing = [8,5,7,8,13]

plt.plot([],[],color='m', label='Sleeping', linewidth=5)
plt.plot([],[],color='c', label='Eating', linewidth=5)
plt.plot([],[],color='r', label='Working', linewidth=5)
plt.plot([],[],color='k', label='Playing', linewidth=5)

plt.stackplot(days, sleeping,eating,working,playing, colors=['m','c','r','k'])
plt.xlabel('Days')
plt.ylabel('Activity')
plt.title('Stack Plot')
plt.legend()
plt.show()
```

```python
names = ['Suresh', 'Ramesh', 'Devesh']
ages = [32, 40, 70]
plt.figure(figsize=(9, 3))
plt.subplot(131)
plt.bar(names,ages)
plt.title('Bar Graph')
plt.subplot(132)
plt.scatter(names, ages)
plt.title('Scatter')
plt.subplot(133)
plt.plot(names, ages)
plt.title('Line Graph')
#plt.suptitle('Age Plots')
plt.show()
```

Histogram Plotting

```python
mu, sigma = 100, 15
x = mu + sigma * np.random.randn(10000)
# the histogram of the data
n, bins, patches = plt.hist(x, 50, density=1, facecolor='g', alpha=0.75) # 50 bins
plt.xlabel('Smarts')
plt.ylabel('Probability')
plt.title('Histogram of IQ')
plt.text(60, .025, r'$\mu=100,\ \sigma=15$')
plt.axis([40, 160, 0, 0.03]) # xmin,xmax,ymon,ymax
plt.grid(True)
plt.show()
```
